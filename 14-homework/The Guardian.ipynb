```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/order-of-australia.csv")
df.head()
```

```python
df['Female'] = 100 - df.percent_male
df['Male'] = df['percent_male']
```

```python
df.head()
```

```python
ax = df.sort_values(by='field', ascending=False).plot(x='field', y=['Male', 'Female'], kind='barh', width=0.4, \
                                                 figsize=(10, 20), color=['steelblue', 'deepskyblue'])
ax.xaxis.tick_top()
```
